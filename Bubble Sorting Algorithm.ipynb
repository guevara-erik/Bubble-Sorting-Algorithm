<h1>Bubble Sorting Algorithm</h1> 

- **Problem to solve:** Sort integers using the bubble sorting algorithm based on the user's input. 
- **Example:** User input [5, 1, 4, 2, 8], system output [1, 2, 4, 5, 8]
    
**Data Information:** 
   - The script will accept integers until a non-integer is typed.

```python
# Stage 1, Collecting the numbers
# try block to handle the exception

try:
    num = []
      
    while True:
        num.append(int(input()))
          
# If the input is not-integer, stop collecting numbers and print the list

except:
    print("\n",num)

#Stage 2, iterations to compare which number is greater and order it accordingly removing and inserting in the correct position
    
for i in range(len(num)-1): 
                            
    print("\n""Pass #",i+1)
    i=0
    
    while i < (len(num)-1): 
       
        if num[i]>num[i+1]:
                
            num.insert(i, num.pop(i+1)) 
            print("Reorder IF",i+1,num)
            i = i+1

        else:
                
            print('Reorder ELSE',i+1,num)
            i = i+1
            
    else:
        print("The new order is",num)  

print("\n""The final order is",num)
```
